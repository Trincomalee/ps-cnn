```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive_output, widgets, VBox, Label

def plot_perceptron(w1_1, w2_1, b_1, w1_2, w2_2, b_2, w1_3, w2_3, b_3, w1_4, w2_4, b_4):
    x1 = np.linspace(-10, 10, 400)
    x2 = np.linspace(-10, 10, 400)
    X1, X2 = np.meshgrid(x1, x2)
    
    # 计算每个网格点的logit值
    logit_1 = w1_1 * X1 + w2_1 * X2 + b_1
    logit_2 = w1_2 * X1 + w2_2 * X2 + b_2
    logit_3 = w1_3 * X1 + w2_3 * X2 + b_3
    logit_4 = w1_4 * X1 + w2_4 * X2 + b_4
    
    # 创建一个布尔数组，用于标记满足所有logit都为正的点
    positive_region = (logit_1 >= 0) & (logit_2 >= 0) & (logit_3 >= 0) & (logit_4 >= 0)
    
    fig, ax = plt.subplots(figsize=(5, 5))
    
    # 绘制4条直线
    x1_plot = np.linspace(-10, 10, 100)
    if w2_1 == 0:
        w2_1 = 1e-10
    if w2_2 == 0:
        w2_2 = 1e-10
    if w2_3 == 0:
        w2_3 = 1e-10
    if w2_4 == 0:
        w2_4 = 1e-10
    
    x2_1 = -(w1_1 * x1_plot + b_1) / w2_1
    x2_2 = -(w1_2 * x1_plot + b_2) / w2_2
    x2_3 = -(w1_3 * x1_plot + b_3) / w2_3
    x2_4 = -(w1_4 * x1_plot + b_4) / w2_4
    
    ax.plot(x1_plot, x2_1, '-b', label='Gerade 1')  
    ax.plot(x1_plot, x2_2, '-g', label='Gerade 2')  
    ax.plot(x1_plot, x2_3, '-r', label='Gerade 3')
    ax.plot(x1_plot, x2_4, '-m', label='Gerade 4')

    # 绘制阴影区域，即这些直线的公共正区域
    ax.contourf(X1, X2, positive_region, levels=[0.5, 1], colors=['red'], alpha=0.1)
    
    # 4个固定的点
    points = np.array([[-5, -5], [-5, 5], [5, -5], [5, 5]])
    colors = []
    coordinates_labels = []
    logit_labels = []
    markers = []
    linewidths = []

    # 计算4个固定的点的logit值并确定颜色和符号
    for (px, py) in points:
        logit_1 = w1_1 * px + w2_1 * py + b_1
        logit_2 = w1_2 * px + w2_2 * py + b_2
        logit_3 = w1_3 * px + w2_3 * py + b_3
        logit_4 = w1_4 * px + w2_4 * py + b_4
        if logit_1 >= 0 and logit_2 >= 0 and logit_3 >= 0 and logit_4 >= 0:
            color = 'red'
            marker = '+'
            linewidth = 2
        else:
            color = 'green'
            marker = 'o'
            linewidth = 1
        colors.append(color)
        markers.append(marker)
        linewidths.append(linewidth)
        coordinates_labels.append(f'({px}, {py})')
        logit_labels.append(f'Logit1: {logit_1:.2f}\nLogit2: {logit_2:.2f}\nLogit3: {logit_3:.2f}\nLogit4: {logit_4:.2f}')
    
    # 绘制固定点和标签
    for idx, (px, py) in enumerate(points):
        ax.plot(px, py, marker=markers[idx], color=colors[idx], markersize=10, markeredgewidth=linewidths[idx])
        ax.text(px, py + 0.5, coordinates_labels[idx], fontsize=9, verticalalignment='bottom', horizontalalignment='center')
        ax.text(px, py - 0.5, logit_labels[idx], fontsize=9, verticalalignment='top', horizontalalignment='center')

    ax.set_title('4 Neuronen, verbunden mit UND-Operation')
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_xlim([-10, 10])
    ax.set_ylim([-10, 10])
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.grid(True)
    ax.legend(loc='upper center')
    plt.show()



# 创建滑动条
w1_1_slider = widgets.FloatSlider(value=-0.8, min=-2.0, max=2.0, step=0.1, description='w1_1')
w2_1_slider = widgets.FloatSlider(value=-1, min=-2.0, max=2.0, step=0.1, description='w2_1')
b_1_slider = widgets.FloatSlider(value=2, min=-10.0, max=10.0, step=0.1, description='b_1')

w1_2_slider = widgets.FloatSlider(value=0.8, min=-2.0, max=2.0, step=0.1, description='w1_2')
w2_2_slider = widgets.FloatSlider(value=-1, min=-2.0, max=2.0, step=0.1, description='w2_2')
b_2_slider = widgets.FloatSlider(value=2, min=-10.0, max=10.0, step=0.1, description='b_2')

w1_3_slider = widgets.FloatSlider(value=-0.5, min=-2.0, max=2.0, step=0.1, description='w1_3')
w2_3_slider = widgets.FloatSlider(value=0.1, min=-2.0, max=2.0, step=0.1, description='w2_3')
b_3_slider = widgets.FloatSlider(value=3.2, min=-10.0, max=10.0, step=0.1, description='b_3')

w1_4_slider = widgets.FloatSlider(value=0.5, min=-2.0, max=2.0, step=0.1, description='w1_4')
w2_4_slider = widgets.FloatSlider(value=0.1, min=-2.0, max=2.0, step=0.1, description='w2_4')
b_4_slider = widgets.FloatSlider(value=3.2, min=-10.0, max=10.0, step=0.1, description='b_4')



# 创建包含滑动条和标签的VBox
ui = VBox([
    Label('Gerade 1'),
    w1_1_slider,
    w2_1_slider,
    b_1_slider,
    Label(' '),
    Label('Gerade 2'),
    w1_2_slider,
    w2_2_slider,
    b_2_slider,
    Label(' '),
    Label('Gerade 3'),
    w1_3_slider,
    w2_3_slider,
    b_3_slider,
    Label(' '),
    Label('Gerade 4'),
    w1_4_slider,
    w2_4_slider,
    b_4_slider
])

# 创建交互式绘图（不在VBox中包含滑动条）
interactive_plot = interactive_output(plot_perceptron, 
                                     {'w1_1': w1_1_slider,
                                      'w2_1': w2_1_slider,
                                      'b_1': b_1_slider,
                                      'w1_2': w1_2_slider,
                                      'w2_2': w2_2_slider,
                                      'b_2': b_2_slider,
                                      'w1_3': w1_3_slider,
                                      'w2_3': w2_3_slider,
                                      'b_3': b_3_slider,
                                      'w1_4': w1_4_slider,
                                      'w2_4': w2_4_slider,
                                      'b_4': b_4_slider})

# 显示滑动条和绘图
display(VBox([interactive_plot, ui]))
```
